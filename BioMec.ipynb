```python
import sys
import os
from sys import platform
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from IPython import display
import tensorflow as tf
%matplotlib inline

sample_video = cv.VideoCapture("./data/video/sample_video.mp4")

ataque = "G7A10At1"
C1_video = cv.VideoCapture("./data/video/ATAQUE1/"+ataque+"C1"+".avi")
##C2_video = cv.VideoCapture("./data/video/ATAQUE1/"+ataque+"C2"+".avi")
##C3_video = cv.VideoCapture("./data/video/ATAQUE1/"+ataque+"C3"+".avi")
##C4_video = cv.VideoCapture("./data/video/ATAQUE1/"+ataque+"C4"+".avi")

def ShowVideo(videoToShow):
    if not videoToShow.isOpened():
        print("Can't get Video")
        return
    
    while(videoToShow.isOpened()):
        try:
            ret, frame = videoToShow.read()
            frame = cv.cvtColor(frame, cv.COLOR_BGR2RGB)
            plt.imshow(frame)

            display.clear_output(wait=True)
            display.display(plt.gcf())

        except KeyboardInterrupt:
            videoToShow.release()
def ExtractVideoData(videoToExtract):
    if not videoToExtract.isOpened():
        print("Can't get Video")
        return
    videoData = {
        "length" : int(videoToExtract.get(cv.CAP_PROP_FRAME_COUNT)),
        "width" : int(videoToExtract.get(cv.CAP_PROP_FRAME_WIDTH)),
        "height" : int(videoToExtract.get(cv.CAP_PROP_FRAME_HEIGHT))
    }
    return videoData
ShowVideo(C1_video)
ExtractVideoData(C1_video)
```
